```python
from hrcsentinel import hrccore as hrc
import matplotlib.pyplot as plt

from astropy.io import fits

import os

%matplotlib notebook
```

```python
full_telem = 'telemfull/'
telem_5min = 'telem5min/'

hrc.styleplots()

%config InlineBackend.figure_format = 'retina'
```

```python
times_2CO5PALV, volts_2CO5PALV = hrc.parse_generic_msid(full_telem + "2C05PALV.csv", valtype="vals")
times_2C15NALV, volts_2C15NALV = hrc.parse_generic_msid(full_telem + "2C15NALV.csv", valtype="vals")
times_2C15PALV, volts_2C15PALV = hrc.parse_generic_msid(full_telem + "2C15PALV.csv", valtype="vals")


times_2CO5PALV, volts_2CO5PALV = hrc.parse_generic_msid(telem_5min + "2C05PALV.csv", valtype="midvals")
times_2C15NALV, volts_2C15NALV = hrc.parse_generic_msid(telem_5min + "2C15NALV.csv", valtype="midvals")
times_2C15PALV, volts_2C15PALV = hrc.parse_generic_msid(telem_5min + "2C15PALV.csv", valtype="midvals")
```

```python
fig, ax = plt.subplots(figsize=(12,8))
ax.plot_date(times_2CO5PALV, volts_2CO5PALV, label='+5V bus monitor (2C05PALV)')
ax.plot_date(times_2C15NALV, volts_2C15NALV, label='-15V bus monitor (2C15NALV)')
ax.plot_date(times_2C15PALV, volts_2C15PALV, label='+15V bus monitor (2C15PALV)')



ax.legend()

ax.set_ylabel('Volts')
ax.set_xlabel('Date')

ax.set_title('Latest telemetry from Ska archive')
```

```python
fig, ax = plt.subplots(figsize=(20,8))

times = hrc.convert_chandra_time(times_2CO5PALV)

ax.plot_date(times, volts_2CO5PALV, label='+5V bus monitor (2C05PALV)')
ax.plot_date(times, volts_2C15NALV, label='-15V bus monitor (2C15NALV)')
ax.plot_date(times, volts_2C15PALV, label='+15V bus monitor (2C15PALV)')



ax.legend()

ax.set_ylabel('Volts')
```
